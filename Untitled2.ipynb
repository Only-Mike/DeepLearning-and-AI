```python
!pip install -U transformers huggingface_hub -qq
!pip install datasets

import pandas as pd
import torch
from transformers import pipeline
from transformers import AutoTokenizer
from transformers import AutoModelForSeq2SeqLM, TrainingArguments, Trainer
from datasets import load_dataset
```

```python
dataset = load_dataset('Rami/adhd_question')
```

```python
#dataset['train'][0]
```

```python
#create test, train and context
train_df = pd.DataFrame(dataset['train'])
test_df = pd.DataFrame(dataset['test'])
```

```python
train_df.head()
```

```python
train_df['context'] = train_df['content']
test_df['context'] = test_df['content']
```

```python
tokenizer = AutoTokenizer.from_pretrained("facebook/blenderbot-1B-distill")
if tokenizer.pad_token is None:
    tokenizer.add_special_tokens({'pad_token': '[PAD]'})

train_encodings = tokenizer(train_df["context"].tolist(), 
                            train_df["Question"].tolist(), 
                            padding=True, 
                            truncation=True)
test_encodings = tokenizer(test_df["context"].tolist(), 
                           test_df["Question"].tolist(), 
                           padding=True, 
                           truncation=True)
```

```python
#Create a PyTorch dataset using the encodings.
class ADHDQuestionDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx])
        return item

    def __len__(self):
        return len(self.labels)

train_dataset = ADHDQuestionDataset(train_encodings, train_df["Answer"].tolist())
test_dataset = ADHDQuestionDataset(test_encodings, test_df["Answer"].tolist())
```

```python
#Instantiate the model and create the Trainer object.
model = AutoModelForSeq2SeqLM.from_pretrained("facebook/blenderbot-1B-distill")

training_args = TrainingArguments(
    output_dir='./results',
    evaluation_strategy='steps',
    eval_steps=1000,
    per_device_train_batch_size=1,
    per_device_eval_batch_size=1,
    num_train_epochs=1,
    save_steps=1000,
    save_total_limit=2
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset
)

#Train the model using the Trainer object.
trainer.train()

```

```python
#Use the trained model for chatbot or conversational bot by running inference using the AutoModelForSeq2SeqLM and AutoTokenizer.
tokenizer = AutoTokenizer.from_pretrained("facebook/blenderbot-1B-distill")
if tokenizer.pad_token is None:
    tokenizer.add_special_tokens({'pad_token': '[PAD]'})

model = AutoModelForSeq2SeqLM.from_pretrained("./results/checkpoint-<insert checkpoint number>/")
```

```python
# Use the model to generate responses to user inputs
def generate_response(user_input):
    input_ids = tokenizer.encode(user_input, return_tensors='pt')
    output = model.generate(input_ids)
    response = tokenizer.decode(output[0], skip_special_tokens=True)
    return response
```
